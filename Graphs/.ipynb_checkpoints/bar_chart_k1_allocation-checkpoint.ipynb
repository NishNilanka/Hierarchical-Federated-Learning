```python
import matplotlib.pyplot as plt

# Data
categories = [
    "Uniform K1 Allocation",
    "High K1 for High Energy",
    "High K1 for Low Energy"
]
values = [6.4097349565, 7.2091276213999995, 5.6091276213999995]

# Colors for a professional look
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=colors, width=0.5)
plt.xlabel("Configurations", fontsize=12)
plt.ylabel("Energy Computation (J)", fontsize=12)
plt.title("Energy Computation for Different Configurations", fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Show the plot
plt.show()
```
