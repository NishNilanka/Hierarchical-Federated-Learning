```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "cleaned_experiment_data.csv"
df = pd.read_csv(file_path)

# Map original k1_allocation labels to more descriptive scientific ones
allocation_labels = {
    "reversed": "High K1 for Low Energy",
    "non-reversed": "High K1 for High Energy",
    "uniform": "Uniform K1 Distribution"
}
df['k1_allocation_label'] = df['k1_allocation'].map(allocation_labels)

# Set Seaborn style
sns.set(style="whitegrid")
scientific_palette = sns.color_palette("colorblind")

# Create the plot
fig, ax = plt.subplots(figsize=(12, 7))
scatter = sns.scatterplot(
    data=df,
    x='average_energy_for_clusters',
    y='local_iterations_for_clusters',
    hue='k1_allocation_label',
    style='k1_allocation_label',
    s=120,
    palette=scientific_palette,
    edgecolor='black',
    ax=ax
)

# Set titles and labels
ax.set_title("Cluster-wise k1 Allocation Strategy", fontsize=16, pad=40)
ax.set_xlabel("Average Energy Consumption per Cluster", fontsize=12)
ax.set_ylabel("Number of Local Iterations Assigned (k1)", fontsize=12)

# Legend positioning and formatting
ax.legend(
    title_fontsize=11,
    fontsize=10,
    bbox_to_anchor=(0.5, 1.09),
    loc="upper center",
    ncol=3,
    frameon=True
)

plt.grid(True)
plt.tight_layout()
plt.savefig("plots/cluster-wise_k1_allocation.png", dpi=600, bbox_inches='tight')
plt.show()
```
